```python
import math
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
    result = math.sqrt(4 - math.sin(x)**2)
    return result
def f2(x):
    if x == 0:
        result=0
    else: result = math.sin(x)/x
    return result
def f3(x):
    result = math.exp(x)/(4+x**2)
    return result
def f4(x):
    result = math.log(x + 1)/(1+x**2)
    return result
```

```python
#复化梯形
def TrapeziumComplex(a,b,epsilon,fStr):
    def F(x):
        if fStr== 'f1':
            return f1(x)
        if fStr== 'f2':
            return f2(x)
        if fStr== 'f3':
            return f3(x)
        if fStr== 'f4':
            return f4(x)
    n = 0
    h0 = b - a
    T0 = h0 / 2 * (F(a) + F(b))
    while True:
        n += 1
        h1 = h0 / 2
        m = 2**(n - 1)
        sum = 0
        for k in range(1,m + 1):
            sum += F(a + (2 * k - 1) * h1)
        T1 = T0 / 2 + h1 * sum
        if abs(T1 - T0) < epsilon:
            break
        T0 = T1
        h0 = h1
    return T1,n
#复化辛普森
def SimpsonComplex(a,b,epsilon,fStr):
    def F(x):
        if fStr== 'f1':
            return f1(x)
        if fStr== 'f2':
            return f2(x)
        if fStr== 'f3':
            return f3(x)
        if fStr== 'f4':
            return f4(x)
    n = 0
    h0 = (b - a) / 2
    S0 = h0 / 3 * (F(a) + 4 * F((a + b) / 2) + F(b))
    while True:
        n += 1
        h1 = h0 / 2
        m = 2**(n - 1)
        sum1 = 0
        sum2 = 0
        for k in range(1,m * 2 + 1):
            sum1 += F(a + (2 * k - 1) * h1)
            if k <= m:
                sum2 += F(a + (4 * k - 2) * h1)
        S1 = S0 / 2 + h1 / 3 * (4 * sum1 - 2 * sum2)
        if abs(S1 - S0) < epsilon:
            break
        S0 = S1
        h0 = h1
    return S1,n
def TwoMethods(a,b,espilon,fStr):
    result1, result2 = TrapeziumComplex(a,b,epsilon,fStr)
    result3, result4 = SimpsonComplex(a,b,epsilon,fStr)
    print("对于下界",a,"上界",b,'函数',fStr)
    print("复化梯形的求积答案为:", result1,"步数为",result2,"步长为",(b-a)/result4)
    print("复化辛普森的求积答案为:", result3,"步数为",result4,"步长为",(b-a)/result4)
    print("·········无情分割线·········")
```
